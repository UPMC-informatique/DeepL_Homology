```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline  
```

## TRAIN SET

```python
f = open("datasets/train_dataset.list.labels_entiers", "r")
tab = []
for line in f:
    l = line.split('\t')
    tab.append(int(l[1]))
```

```python
tab = np.sort(tab)

mini=0
maxi = 1600

nb = int((maxi-mini)/200)
vecteur = np.arange(nb)
vecteur_labels=[]
val = mini
val2 = mini + 200
total = 0
for i in range (0, nb):
    for j in tab:
        if((j>=val)&(j<=val2)):
            total = total + 1
    vecteur[i] = total
    vecteur_labels.append(str(val)+"-"+str(val2))
    val = val + 200
    val2 = val2 + 200

plt.rcdefaults()
fig, ax = plt.subplots()
y_pos = np.arange(nb)

ax.barh(y_pos, vecteur, align='center',
        color='green', ecolor='black')

ax.set_yticks(y_pos)
ax.set_yticklabels(vecteur_labels)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('nombre de proteines')
ax.set_title('bins')

plt.show()

print(vecteur)
```
